```python
import numpy as np
n = int(input())
a = np.array([int(input())for i in range(n)])
b = np.array([int(input())for i in range(n)])

# In ra max, min, giá trị trung bình, độ lệch chuẩn của vector, giá trị trung vị
print(f"max of array a = {np.max(a)}")
print(f"min of array a = {np.min(a)}")
print(f"mean of array a = {np.mean(a)}")
print(f"standard of array a = {np.std(a)} ")
print(f"median of array a = {np.median(a)}")
print()
print(f"max of array b = {np.max(b)}")
print(f"min of array b = {np.min(b)}")
print(f"mean of array b = {np.mean(b)}")
print(f"standard of array b = {np.std(b)} ")
print(f"median of array b = {np.median(b)}")

#Tạo 1 ma trận numpy c (2-D) có kích thước nx2 từ a, b, in ra màn hình.
c = np.array([a,b]).T
print(c)
print()

#Khởi tạo 1 ma trận (2-D) numpy d với kích thước nxn theo phân phối Gauss (với mean, std của b), in kết quả ra màn hình.
meanB = np.mean(b)
stdB = np.std(b)
d = np.random.normal(meanB,stdB,(n,n))
print(d)
print()

# Tính tổng, hiệu, tích 2 ma trận, tích ma trận chuyển vị của d với ma trận nghịch đảo của d
print(f"sum = {a+b}")
print(f"sub = {a-b}")
print(f"mul1  = {a * b}10")
print(f"mul2 = {np.dot(d.T,np.linalg.inv(d))}")
print()

#Chuyển ma trận c về tensor numpy e (3-D) (nếu không đủ dữ liệu để chuyển về 3-D thì thêm dữ liệu), in ra màn hình.
e = np.expand_dims(c,axis = 0)
print(e)



```
