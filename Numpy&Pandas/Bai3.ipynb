```python
import numpy as np
# Lập 1 mảng 2 chiều lưu lần lượt điểm đại số tuyến tính, xác suất thống kê, giải tích với 20 phần tử mỗi hàng random trong khoảng từ [0, 10].
a = np.random.randint(0,10,(3,20))
print(a)
print()
#In ra giá vị trí điểm lớn nhất và nhỏ nhất ở mỗi môn.
print(f"index of max = {np.argmax(a)}")
print(f"index of min = {np.argmin(a)}")
print()

#Làm phẳng mảng và xóa đi các vị trí có điểm là 0, in ra màn hình.
b = a.flatten()[a.flatten() != 0]
print(b)
print()
#Sắp xếp giảm theo thuật toán quicksort, in ra màn hình.
c = np.sort(b)[::-1]
print(c)
print()
#
k = int(input("nhap k: "))
insert_pos = np.searchsorted(b,k,side = 'right')
d = np.insert(b,insert_pos,k)
print(d)





```
