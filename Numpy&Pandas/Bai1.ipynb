```python
import numpy as np
#Khởi tạo một mảng numpy nguyên a (1-D) với n phần tử ngẫu nhiên trong khoảng từ [1, 14]
n = int(input())
a = np.random.randint(1,15,n)
print(a)
print("dim: ", a.ndim, " ,shape: ", a.shape, " ,len: ",len(a), " ,itemsize: ",a.itemsize, " ,dtype: ",a.dtype)
print()

#Khởi tạo 1 mảng numpy b với 100 số chẵn đầu tiên.
b = np.arange(0,200,2)
print(b)
print("dim: ", b.ndim, " ,shape: ", b.shape, " ,len: ",len(b), " ,itemsize: ",b.itemsize, " ,dtype: ",b.dtype)
print()

#Khởi tạo 1 ma trận numpy c có kích thước nxn với 100 phần tử 0, in ra màn hình.
c = np.zeros((10,10))
print(c)
print("dim: ", c.ndim, " ,shape: ", c.shape, " ,len: ",len(c), " ,itemsize: ",c.itemsize, " ,dtype: ",c.dtype)
print()

#Khởi tạo ma trận đơn vị d có kích thước nxn.
d= np.eye(n)
print(d)
print("dim: ", d.ndim, " ,shape: ", d.shape, " ,len: ",len(d), " ,itemsize: ",d.itemsize, " ,dtype: ",d.dtype)
print()

#Khởi tạo ma trận đường chéo e với các giá trị trên đường chéo là mảng a.
e=np.diag(a)
print(e)
print("dim: ", e.ndim, " ,shape: ", e.shape, " ,len: ",len(e), " ,itemsize: ",e.itemsize, " ,dtype: ",e.dtype)




```
