```python
#Bài 1

import math
xA = int(input())
yA = int(input())
xB = int(input())
yB = int(input())
xC = int(input())
yC = int(input())

ab = math.sqrt((xB - xA)**2 + (yB - yA)**2)
ac = math.sqrt((xC - xA)**2 + (yC - yA)**2)
bc = math.sqrt((xC - xB)**2 + (yC - yB)**2)

if(ab + ac > bc and ab + bc > ac and ac + bc > ab):
    print('TAM GIÁC')
else:
    print('KHÔNG PHẢI TAM GIÁC')
```

```python
#Bài 2

def reverse(number, m):
    n = 0
    tmp = 1
    x = 0

    for i in range(m):
        tmp *= 10
        x = x*10 +( number % 10)
        number = number // 10
    num = number

    result = num * tmp  + x
    return result


n = int(input())
m = int(input())

for i in range(n):
    number = int(input())
    result = reverse(number, m)
    print(result)
```

```python
#Bài 3

k = int(input())
for i in range(k):
    name = input()
    p1 = int(input())
    p2 = int(input())
    sum = p1 +p2
    if(sum >= 190):
        print(i+1, name, sum, 'Xuất sắc\n')
    elif(sum >=150):
        print(i+1,name,sum,'Giỏi\n')
    elif(sum >= 100):
        print(i+1,name,sum,'Khá\n')
    else:
        print(i+1,name,sum,'Yếu\n')
```
