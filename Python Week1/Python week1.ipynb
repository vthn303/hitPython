*   Cách khai báo biến
> Python có cách khai báo biến đơn giản, với cấu trúc
```
ten_bien = gia_tri
```
Ví dụ:
```
name = 'Vu Thi Hong Nhung'     #Khai báo biến name với giá trị chuỗi là chuỗi "Vu Thi Hong Nhung"
age = 19     #Khai báo biến age với giá trị là số nguyên 19
weight = 54.6     #Khai báo biến weight với giá trị là số thực 54.6
```
Quy tắc khai báo biến trong Python
>*   Tên biến chỉ có thể chứa chữ cái, số và dấu gạch dưới (_). Có thể bắt đầu bằng chữ cái hoặc dấu gạch dưới, nhưng không được bắt đầu bằng số.
>*   Tên biến phải phân biệt chữ hoa và chữ thường. Ví dụ: name và Name là hai biến khác nhau.
>*   Tên biến không được trùng với các từ khóa của Python. Các từ khóa là các từ có ý nghĩa đặc biệt trong ngôn ngữ, ví dụ: if, for, while, def, class, …
>*   Tên biến nên có ý nghĩa và dễ hiểu. Không nên đặt tên biến quá ngắn hoặc quá dài.

* Các toán tử
>* Toán tử số học


````python
#Khai báo biến
a = 5.1
b = 6.2


#Phép cộng hai số
print('a + b = ', (a + b))

#Phép trừ hai số
print('a - b = ', (a - b))

#Phép nhân hai số
print("a * b = ", (a * b))

#Phép chia hai số
print("a / b = ", (a / b))

#Phép chia lấy dư hai số
print('a % b = ', (a % b))

#Lũy thừa
print('a^b = ', (a ** b))
````


>*   Toán tử quan hệ: dùng để so sánh giá trị của hai toán hạng và trả về giá trị True hoặc False như ==, !=, >, <, >=, <=,

````python
a = 5
b = 10


print(a == b)
print(a != b)
print(a > b)
print(a < b)
````



*   Toán tử logic: dùng để thực hiện các phép toán logic gồm có:
>*   Toán tử ***and*** dùng thực hiện phép toán logic ***và***
>*   Toán tử ***or*** dùng thực hiện phép toán logic ***hoặc***
>*   Toán tử ***not*** dùng thực hiện phép toán logic ***phủ định***


````python
a = True
b = False

print(a and b)
print(a or b)
print(not a)
````



*   Toán tử gán giá trị: dùng để thiết lập giá trị ban đầu hoặc thay đổi giá trị của biến


````python
a = 5
b = 10

a += b
print(a)

a -= b
print(a)

a *= b
print(a)

a /= b
print(a)

````

* Toán tử Membership: dùng để kiểm tra sự tồn tại của một giá trị trong một giá trị cho trước. Giá trị cho trước này thường có kiểu dữ liệu là string (chuỗi),list (danh sách), tuple… Python có hai toán tử membership là ***in*** và ***not in*** (ngược lại của ***in***).

````python
a = "Xin Chao!"

print('X' in a)
print('C' not in a)
print("!" in a)
````



*   Toán tử Identify: dùng để kiểm tra xem hai biến có cùng tham chiếu đến một đối tượng hay không


````python
x = 20
y = 20
if (x is y):
    print("x & y  giống nhau")
y = 30
if (x is not y):
    print("x & y khác nhau")
````



*   Các kiểu dữ liệu
>* Numbers
>>* Các kiểu dữ liệu số trong Python được phân loại bằng các từ khóa sau: int, float và complex (số phức)
>>*Python có một hàm có sẵn là  type()  để xác định kiểu dữ liệu của một biến hoặc giá trị.
>>*Một hàm có sẵn khác là isinstance() để kiểm tra loại đối tượng.
>>*Trong Python, chúng ta có thể thêm “j” hoặc “J” sau một số để quy ước số phức.

```
a = 2
print(type(a))    #ketqua: <class 'int'>

a = 3.0
print(type(a))    #ketqua: <class 'float'>

a = 2 + 3j
print(type(a))    #ketqua: <class 'complex'>

```
>* Strings
>>* Chuỗi kí tự là chuỗi các kí tự Unicode
>>*Có thể sử dụng dấu nháy đơn '' hoặc dấu nháy kép "" để biểu diễn chuỗi ký tự
>>*Chuỗi kí tự trên nhiều dòng có thể dùng dấu nháy kép, hoặc 3 dấu nháy kép, hoặc 3 dấu nháy đơn

```
str = 'hello'
print(str)
#ketqua: hello

str = "hi"
print(str)
#ketqua: hi

str = """hello
every
one!"""
print(str)
#ketqua: hello
every
one!
```
>* Lists
>>* Là một mảng lưu trữ dữ liệu
>>* Mỗi phần tử trong mảng có thể lưu trữ những dữ liệu khác nhau
>>* Nó rất linh hoạt và không có kích thước cố định
>>* Giá trị các phần tử trong tập hợp có thể thay đổi được

```
list = [10, 20.5, 'hello']
print('list[0] = ', list[0])
#ketqua: 10

print('list[0:2] = ', list[0:2])
#ketqua: [10, 20.5]

print('list[2:] = ', list[2:])
#ketqua: ['hello']

```

>* Tulpe
>>* Là một chuỗi các phần tử có thứ tự
>>* Các giá trị của các phần tử là không thể thay đổi. Các tập hợp kiểu Tuple sau khi được tạo, ta sẽ không thể sửa đổi giá trị của nó.

```
tuple = (10, 10.2, 'hi')

print('tuple[0] = ', tuple[0])
#ketqua: 10

print('tuple[0:2] = ', tuple[0:2])
#ketqua: (10, 10.2)

```
>* Sets
>>* Là một tập hợp không có thứ tự
>>* Các phần tử trong set không được sắp xếp
>>* Các đối tượng kiểu dữ liệu trong Set có các giá trị duy nhất và không lặp lại

```
set = {3, 0, 6, 2, 6, 10, 10}

print(set)
#ketqua: {0, 2, 3, 6, 10}
```
>* Dictionaries
>>* Là một tập hợp các cặp khóa và giá trị không có thứ tự
>>* Dictionary được tối ưu hóa để truy xuất một lượng lớn dữ liệu
>>* Mỗi phần tử trong Dictionary là một cặp khóa và giá trị. Khóa và giá trị có thể thuộc bất kì dữ liệu nào

```
dict = {"HP": 100, 'Dame': 20}
print(dict['HP'])
#ketqua: 100
```
